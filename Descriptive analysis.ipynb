```python
import pandas as pd
import re
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df_covariates = pd.read_csv("data/lists of songs/dataframe_song_features.csv")
```

```python
df_covariates.head()
```

```python
df_covariates = df_covariates.drop(df_covariates.columns[[0]],axis = 1)
```

```python
print("Shape"+str(df_covariates.shape))
df_covariates.info()
```

```python
df_songs= pd.read_csv("data/lists of songs/dataframe_with_lyrics_completecases.csv")
```

```python
df_songs.head()
```

```python
df_songs = df_songs[["songs", "artist", "title"]]
```

```python
df_songs.shape
```

```python
df_full = pd.merge(df_songs,df_covariates,how='left',left_on=["songs"], right_on=["songs"])
```

```python
df_full.shape
```

```python
df_full = df_full.drop_duplicates()
```

```python
df_full.head()
```

```python
df_full['year'] = df_full['title'].str.extract('(\d{4})', expand=True)
```

```python
df_full['country'] = df_full['title'].str.extract(r'Top Songs of (\d\d\d\d): (.*)')[1]
```

```python
df_full["country"].value_counts()
```

```python
df_full.columns
```

```python
df_grouped = (
    df_full[['danceability', 'energy',
       'key', 'loudness', 'mode', 'speechiness', 'acousticness',
       'instrumentalness', 'liveness', 'valence', 'tempo', 'title']].groupby(['title'])
    .agg(['mean'])
)
```

```python
df_grouped
```

```python
df_full.to_csv("data/lists of songs/full_dataset.csv")
```

```python
df_grouped.to_csv("data/lists of songs/summary.csv")
```
