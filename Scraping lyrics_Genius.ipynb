```python
# To identify cases where lyrics are not found, we find the fields with enumerated lists

def find_enumerated_list(string):
    regex = re.compile(r'(\d+\.)(.*?)(?=\d+\.|$)', re.DOTALL)
    matches = regex.finditer(string)
    matches_list = [match.group() for match in matches]
    return matches_list
```

```python
df['enumerated_list'] = df['lyrics'].apply(find_enumerated_list)
```

```python
not_found_lyrics = df[df['enumerated_list'].apply(len) > 0]
```
